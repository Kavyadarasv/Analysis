```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
movies_df=pd.read_csv('DS1_C8_V3_ND_Sprint2_Data Analysis Using Python_Dataset.csv')
movies_df
```

```python
import json
```

```python
movie_data=movies_df.copy()
```

```python
columns_dict  = ['genres','keywords','production_companies','production_countries','spoken_languages']
```

```python
for column in columns_dict:
    movie_data[column] = movie_data[column].apply(lambda x: [item ['name'] for item in json.loads(x)])
```

```python
movie_data
```

```python
columns_to_extract = ['genres','keywords','production_companies','production_countries','spoken_languages']
```

```python
for col in columns_to_extract:
    movie_data[col]=movie_data[col].apply(lambda x:x[0] if isinstance (x,list) and x else None)
    
# ifinstance(x,list) and 'x' checks if the cell contains a list and not empty,it extracts the first element 
# using x[0]. if not it assigns 'None' to that cell   
```
